```python
#make an array containing all the transactions
d=[]
for i in range(len(df)) : 
    k=[]
    for j in df.iloc[i, :]:
        if(type(j)==str):
            
            k.append(j)
    d.append(k)
```

```python
#create a dictionary to store the number of times each item is bought 
count={};
for i in d:
    for j in i:
        count[j]=count.get(j,0)+1;
count
```

```python
#remove items with frequency less than minimum support of 40
minimum=40
L=count.keys()
p=[]
final_count={}
for item in L:
    p.append(item)
    if count[item]>minimum:
        final_count[item]=count[item]

final_count=sorted(final_count.items(), key = lambda kv:(kv[1], kv[0]))
final_count.reverse()
final_count
```

```python
#making an ordered dictionary
from collections import OrderedDict
dictionary=OrderedDict()

for t in final_count:
    e=t[0]
    f=t[1]
    dictionary[e]=f
#make a final list q
q=[];
for t in dictionary:
    q.append(t);
q.reverse();
```

```python
#functions to make and update the tree
class node:

    def __init__(self, name, freq):
        self.name = name
        self.freq = freq
        self.children=[]
        self.child={}
        
    def disp(self, ind=1):
        print ('  '*ind, self.name, '  ', self.freq)
        for child in self.child.values():
            child.disp(ind+2)
    
def update(n,k):
    if k in n.children:
        n.child[k].freq+=1;
        
    else:
        n.child[k]=node(k,1)
        n.children.append(k)
```

```python
#start making the tree
null=node('start',1);

n=null;
for t in d:
    n=null;
    for i in q:
        if i in t:
            update(n,i)
            n=n.child[i]
null.disp() 
```
